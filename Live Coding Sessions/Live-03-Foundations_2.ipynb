#### Task 4.3: Converting Another Column

Floats first!

````python
import numpy as np

def to_float(col_data):
    fdata = []
    for c in col_data:
        try:
            fdata.append( float(c) )
        except (ValueError, TypeError):
            fdata.append( np.nan )
    return fdata

float_cols = ['Latitude','Longitude']
for f in float_cols:
    ds[ f ] = to_float( ds[f] )
````
